```python
import re

# 원본 INSERT SQL문 (필요 시 여기에 전체 SQL문 붙여넣기)
original_sql = """

"""

# 정규식 패턴으로 INSERT VALUES 문 중 대여ID 추출 및 수정
def modify_rental_id(sql_text):
    pattern = r"(INSERT INTO 대여_이력\s*\(.*?\)\s*VALUES\s*\()(\d+)(,)"
    modified_sql = re.sub(pattern, lambda m: f"{m.group(1)}{int(m.group(2)) + 1515}{m.group(3)}", sql_text)
    return modified_sql

# 수정된 SQL문 출력
modified_sql = modify_rental_id(original_sql)
print(modified_sql)
```

```python
import re

# 원본 SQL 텍스트를 변수로 저장 (아래는 생략한 예시이며, 실제 전체 SQL을 넣어야 합니다)
sql_text = """


"""

# 정규식으로 각 INSERT 블록을 추출
insert_pattern = r"(INSERT INTO 대여_이력\s*\([^)]*\)\s*VALUES\s*\(\s*)(\d+)(\s*,.*?\);)"
matches = re.findall(insert_pattern, sql_text, re.DOTALL)

# 1515를 더한 새 INSERT 문 생성
new_insert_statements = []
for prefix, 대여ID, suffix in matches:
    new_id = int(대여ID) + 95
    new_stmt = f"{prefix}{new_id}{suffix}"
    new_insert_statements.append(new_stmt)

# 출력
new_sql = "\n".join(new_insert_statements)
print(new_sql)
```
